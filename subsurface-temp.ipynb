# Subsurface lake temperature warming

Recreating [Fig 4 from this paper](https://www.nature.com/articles/s41467-021-21971-1), using data provided to AFP by author.

```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('data/SubsurfaceLakeTemp.csv')
```

```python
df
```

```python
# Make depth the index
df.index = df.Depth
# Remove the original Depth column
df.drop(columns=['Depth'], inplace=True)
```

```python
# Set chart size
sns.set(rc={'figure.figsize':(8,5)})
```

```python
# Plot a heatmap
sns.heatmap(df)
```

```python
# Let's try cool-warm colour palette
sns.color_palette("coolwarm", as_cmap=True)
```

```python
sns.set(rc={'figure.figsize':(10,4)})
```

```python
# Plot with cool-warm colours
sns.heatmap(df, cmap='coolwarm', linewidths=.5, linecolor='lightblue', square=True, cbar_kws={'shrink':.5})\
    .set_title('Deep water warming in lake Michigan', fontsize=16)
```

```python
# Change font & set width/height ratio
sns.set(rc={'figure.figsize':(10,2), 'figure.facecolor':'white'},
        font="Verdana")
```

```python
# Redraw heatmpa
sns.heatmap(df, cmap='coolwarm', linewidths=.5, linecolor='lightblue', cbar_kws={'shrink':.5})\
    .set_title('Deep water warming in lake Michigan', fontsize=14)
```

```python
# Try to match Sabrina's colour suggestions
sns.diverging_palette(240, 60, l=75, s=200, as_cmap=True)
```

```python
# Make this a variable
sab = sns.diverging_palette(240, 60, l=75, s=200, as_cmap=True)
```

```python
# Redraw chart with new colours
sns.heatmap(df, cmap=sab, linewidths=.5, linecolor='white', cbar_kws={'shrink':.5})\
    .set_title('Deep water warming in lake Michigan', fontsize=14)
```

```python
# Save chart as SVG

fig = sns.heatmap(df, cmap=sab, linewidths=.5, linecolor='lightblue', cbar_kws={'shrink':.5})\
    .set_title('Deep water warming in lake Michigan', fontsize=14).get_figure()

fig.savefig('images/Subsurface_temp.svg')
```
