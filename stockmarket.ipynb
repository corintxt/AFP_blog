```python
import matplotlib.pyplot as plt
import pandas as pd
import afp_style as style
plt.style.use('https://storage.googleapis.com/datanode/afpchart.mplstyle')
```

Source: St. Louis Fed [S&P500 index](https://fred.stlouisfed.org/series/SP500)

```python
## Load data
df = pd.read_csv('data/SP500-10yearB.csv')
```

```python
df.tail()
```

```python
df['observation_date'] = pd.to_datetime(df['observation_date'])
```

```python
# Make date the index
df.set_index('observation_date', inplace=True)
```

```python
# Ensure the index is sorted
df.sort_index(inplace=True)

chart = df.plot(
        kind='line',
        color='red')

style.style_plot(plt,
    title='A strong decade for stocks',
    subtitle='US stocks performed better than the historical average in the past 10 years',
    source='S&P Dow Jones Indices, St. Louis Fed')

# plt.savefig('TenYearStocks.svg')
```
