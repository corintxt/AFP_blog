# House prices in Nevada

Source: [Bankrate.com](https://www.bankrate.com/real-estate/housing-market/nv/)

To get the data:

Use Developer Console > Sources > `real-estate/housing-market/nv > nv/` to see source code

Then Ctrl+F 'all-residential', save the JSON data.

```python
import pandas as pd
import json
```

```python
f = open('data/bankrateNVprices.json')
```

```python
data = json.load(f)
```

```python
f.close()
```

```python
# Extract the dataset from the JSON
dataset = data['median_home_values']['dataset']

df = pd.DataFrame(dataset)

df['timestamp'] = pd.to_datetime(df['timestamp']).dt.date
```

```python
df
```

```python
df.plot(x='timestamp', y='value')
```

```python
df.sort_values(by='timestamp',inplace=True)
```
