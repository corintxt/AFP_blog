## RCP Poll Scraper

[Target: Trump vs. Biden](https://www.realclearpolitics.com/epolls/2024/president/us/general-election-trump-vs-biden-7383.html)

```python
import requests
import json
```

```python
URL = 'https://www.realclearpolitics.com/poll/race/7383/polling_data.json'
resp = requests.get(URL)
```

```python
# Check for 200 OK status
print(resp.status_code)
```

```python
# Convert response to json
rjson = json.loads(resp.text)
```

```python
# Look at the moduleInfo data
rjson['moduleInfo']
```

```python
# Poll results are in a big list under the key 'poll'
all_polls = rjson['poll']
```

```python
# Access first list item to look at data format
all_polls[0]
```

```python
# Iterate through each list item and print different dictionary properties
for poll in all_polls:
    print(f"{poll['pollster']} | {poll['date']} | {poll['sampleSize']} | {poll['spread']['name']} {poll['spread']['value'] }")
```
