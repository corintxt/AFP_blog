# Ice cover in Michigan lakes

Source: [GLERL Historical Ice Cover](https://www.glerl.noaa.gov/data/ice/#historical)

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
ice = pd.read_csv('data/AMIC-by-lake.txt', delim_whitespace=True)
```

Columns:

yrs = year of sample

other columns represent ice cover on each lake:
* sup - Superior
* mic - Michigan
* hur - Huron
* eri - Eerie
* ont - Ontario
* bas - Great Lakes basin

```python
ice.head()
```

```python
ice.plot(x='yrs', y='mic')
```

The yearly variation makes it harder to see a trend -- so let's draw a line of best fit showing average cover over time.

Reference: [making a line of best fit in Python](https://www.statology.org/line-of-best-fit-python/)

```python
a, b = np.polyfit(ice.yrs, ice.mic, 1)
```

```python
# Draw chart
plt.figure(figsize=(12,5))
plt.plot(ice.yrs, ice.mic, color='#6DCCD3')
plt.fill_between(ice.yrs, ice.mic) # creates area chart
plt.plot(ice.yrs, a*ice.yrs + b, color='#C0F7FE', linestyle='--')
plt.title('Annual ice coverage on Lake Michigan-Huron', fontdict=None, loc='center', pad=None)
plt.xlabel("Year")
plt.ylabel("Max coverage (percent)")
plt.savefig('images/ice-coverage.svg')
```
